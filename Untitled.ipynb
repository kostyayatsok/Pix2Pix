```python
import matplotlib.pyplot as plt
import glob
from PIL import Image
from random import shuffle
plt.figure(figsize=(10, 10))
paths = glob.glob('predictions_val/*.jpg')
shuffle(paths)
for i, img_p in enumerate(paths):
    plt.subplot(4, 4, i + 1)
    img = Image.open(img_p)
    plt.imshow(img)
    plt.axis('off')
    if i == 15:
        break
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.savefig('example.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import glob
from PIL import Image
from random import shuffle
import numpy as np

plt.figure(figsize=(8, 10))
paths = glob.glob('predictions_val/*.jpg')
shuffle(paths)
for i, p in enumerate(paths):
    fname = p.split('/')[-1]
    
    img_l1 = Image.open(f"predictions_val_l1/{fname}")
    img_gan = Image.open(f"predictions_val_gan/{fname}")
    img = np.array(Image.open(f"facades/val/{fname}"))
    
    
    plt.subplot(5, 4, 4 * i + 1)
    if i == 0: plt.title("mask")
    plt.imshow(img[:,256:,:])
    plt.axis('off')
    plt.subplot(5, 4, 4 * i + 2)
    if i == 0: plt.title("Orig")
    plt.imshow(img[:,:256,:])
    plt.axis('off')
    plt.subplot(5, 4, 4 * i + 3)
    if i == 0: plt.title("L1")
    plt.imshow(img_l1)
    plt.axis('off')
    plt.subplot(5, 4, 4 * i + 4)
    if i == 0: plt.title("L1+cGAN")
    plt.imshow(img_gan)
    plt.axis('off')
    if i == 4:
        break
plt.subplots_adjust(wspace=.05, hspace=.05)
plt.savefig('example_l1.jpg')
plt.show()
```
